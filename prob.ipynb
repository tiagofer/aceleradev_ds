```python
import pandas as pd
import numpy as np 
import scipy.stats as scp 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
norm = scp.norm.rvs(loc=10,scale=3,size=1000)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(norm)
```

```python
scp.norm.cdf(15,loc=10,scale=3)
```

```python
x = np.linspace(0,20,num=1000)
f = lambda x: scp.norm.cdf(x,loc=5,scale=3)
cdf = f(x)
sns.lineplot(x,cdf)
```

```python
scp.norm.sf(8,loc=10,scale=3)
```

```python
bin_20_05 = scp.binom.rvs(20,0.5,size=100000)
binom_20_07 = scp.binom.rvs(20, 0.7, size=100000)
sns.distplot(bin_20_05,bins=range(4, 18), kde=False, hist_kws={"alpha": 0.9})
sns.distplot(binom_20_07, bins=range(8, 22), kde=False, hist_kws={"alpha": 0.5})
```
